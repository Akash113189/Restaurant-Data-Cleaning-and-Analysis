```python
df.to_excel('Cleaned_Resturent_Data.xlsx', index = False) #exporting the cleaned dataset in excel format
```

```python
#Creating a new column to give the Rating to the resturant.
def calculate_rating(review):
    # Split the review by a comma and count the entries
    entry_count = len(review.split(',')) if review else 0
    if entry_count >= 5:
        return 5
    elif entry_count >= 4:
        return 4
    elif entry_count >= 3:
        return 3
    elif entry_count >= 2:
        return 2
    else:
        return 1
```

```python
df["Review Rating"] = df["Review"].apply(calculate_rating)
```

```python
df
```

```python
df = df.sort_values(by="Review Rating", ascending = False)
```
